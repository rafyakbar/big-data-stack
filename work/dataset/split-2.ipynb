```python
import pandas as pd

input_file = "raid-filtered.jsonl"
output_prefix = "raid-filtered_"
chunk_size = 25000  # jumlah baris per file

for i, chunk in enumerate(pd.read_json(input_file, lines=True, chunksize=chunk_size)):
    file_index = f"{i+1:03d}"
    output_file = f"splits-filtered/{output_prefix}{file_index}.jsonl"
    chunk.to_json(output_file, orient="records", lines=True, force_ascii=False)
    print(f"File {output_file} selesai disimpan.")
```
